```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('data.xlsx')

# Step 1: Calculate IQR (Interquartile Range) to identify outliers
Q1 = df['Salary'].quantile(0.25)  # 25th percentile
Q3 = df['Salary'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 2: Separate normal salaries and outliers
normal_salaries = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]['Salary']
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]['Salary']

# Step 3: Plot the histogram
plt.figure(figsize=(10, 6))

# Plot normal salaries
plt.hist(normal_salaries, bins=20, color='blue', alpha=0.7, label='Normal Salary',edgecolor='black')

# Plot outliers
plt.hist(outliers, bins=20, color='red', alpha=0.7, label='Outliers',edgecolor='black')

# Add labels and title
plt.title('Salary Distribution with Outliers Highlighted', fontsize=16)
plt.xlabel('Salary', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('data.xlsx')

# Function to plot histogram with outliers highlighted
def plot_histogram_with_outliers(column_name):
    # Step 1: Calculate IQR to identify outliers
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Step 2: Separate normal values and outliers
    normal_values = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)][column_name]
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)][column_name]

    # Step 3: Plot the histogram
    plt.figure(figsize=(8, 5))

    # Plot normal values
    plt.hist(normal_values, bins=20, color='blue', alpha=0.7, label='Normal ' + column_name, edgecolor='black')

    # Plot outliers
    plt.hist(outliers, bins=20, color='red', alpha=0.7, label='Outliers ' + column_name, edgecolor='black')

    # Add labels and title
    plt.title(f'{column_name} Distribution with Outliers Highlighted', fontsize=16)
    plt.xlabel(column_name, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.legend()

    # Show the plot
    plt.show()

# Plot histograms for each personality trait
plot_histogram_with_outliers('conscientiousness')
plot_histogram_with_outliers('agreeableness')
plot_histogram_with_outliers('extraversion')
plot_histogram_with_outliers('nueroticism')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('data.xlsx')

# Step 1: Calculate the frequency of each designation
designation_counts = df['Designation'].value_counts()

# Step 2: Select the top 20 designations
top_20_designations = designation_counts.head(20)

# Step 3: Plot the bar chart for the top 20 designations
plt.figure(figsize=(12, 6))

# Plotting the bar chart with black borders
bars = plt.bar(top_20_designations.index, top_20_designations.values, color='skyblue', edgecolor='black', linewidth=1.5)

# Add labels and title
plt.title('Top 20 Designations Distribution', fontsize=16)
plt.xlabel('Designation', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Rotate x labels for better visibility
plt.xticks(rotation=45, ha='right')

# Adjust bottom to avoid label cut-off
plt.subplots_adjust(bottom=0.25)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('data.xlsx')

# List of technical assessment columns
technical_columns = ['ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience',
                     'MechanicalEngg', 'ElectricalEngg', 'CivilEngg', 'TelecomEngg']

# Function to plot histogram with outliers highlighted
def plot_histogram_with_outliers(column_name):
    # Step 1: Calculate IQR (Interquartile Range) to identify outliers
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Step 2: Separate normal values and outliers
    normal_values = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)][column_name]
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)][column_name]

    # Step 3: Plot the histogram
    plt.figure(figsize=(10, 6))

    # Plot normal values
    plt.hist(normal_values, bins=20, color='blue', alpha=0.7, label='Normal ' + column_name, edgecolor='black')

    # Plot outliers
    plt.hist(outliers, bins=20, color='red', alpha=0.7, label='Outliers ' + column_name, edgecolor='black')

    # Add labels and title
    plt.title(f'{column_name} Distribution with Outliers Highlighted', fontsize=16)
    plt.xlabel(column_name, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.legend()

    # Show the plot
    plt.show()

# Plot histograms for each technical assessment column
for column in technical_columns:
    plot_histogram_with_outliers(column)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('data.xlsx')

# Step 1: Remove rows with missing values in 'Salary' or 'CollegeState'
df_clean = df[['Salary', 'CollegeState']].dropna()

# Step 2: Calculate the average salary for each CollegeState
avg_salary_by_state = df_clean.groupby('CollegeState')['Salary'].mean().sort_values(ascending=False)

# Step 3: Plot the bar chart
plt.figure(figsize=(12, 6))

# Plotting the bar chart with black borders
bars = plt.bar(avg_salary_by_state.index, avg_salary_by_state.values, color='skyblue', edgecolor='black', linewidth=1.5)

# Add labels and title
plt.title('Average Salary by College State', fontsize=16)
plt.xlabel('College State', fontsize=14)
plt.ylabel('Average Salary', fontsize=14)

# Rotate x labels for better visibility
plt.xticks(rotation=45, ha='right')

# Adjust bottom to avoid label cut-off
plt.subplots_adjust(bottom=0.2)

# Show the plot
plt.show()
```
